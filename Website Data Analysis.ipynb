     time?

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data-export.csv")
df.head()
```

# Remove the unnamed

```python
df.columns = df.iloc[0]
df.head()
```

```python
df = df.drop(index=0).reset_index(drop=True)

df.columns = [
    "channel group", "DateHour", "Users", "Sessions", "Engaged sessions",
    "Average engagement time per session", "Engaged sessions per user",
    "Events per session", "Engagement rate", "Event count"
]

df.head()
```

```python
df.info()
```

# change the date hour

```python
df["DateHour"] = pd.to_datetime(df["DateHour"], format="%Y%m%d%H", errors='coerce')

numeric_cols = df.columns.drop(["channel group", "DateHour"])  

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

df["Hour"] = df["DateHour"].dt.hour
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

# Session and usere over time 

```python
sns.set(style = "whitegrid")
```

```python
plt.figure(figsize=(10,5))
df.groupby("DateHour")[["Sessions","Users"]].sum().plot(ax=plt.gca())
plt.title("Session and usere over time")
plt.xlabel("DateHour")
plt.ylabel("Count")
plt.show()
```

# Total users By Channel

```python

plt.figure(figsize=(12, 7))
sns.barplot(data=df, x="channel group", y="Users", estimator=np.sum, palette="viridis")
plt.title("👥 Total Users by Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

#  Average Engagement Time By Channel

```python
plt.figure(figsize=(9, 7))
sns.barplot(
    data=df,
    x="channel group",
    y="Average engagement time per session",
    estimator=np.mean,
    palette="magma"
)
plt.title("Average Engagement Time by Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

# Engagement Rate Distribution by Channel

```python
plt.figure(figsize=(9, 7))
sns.boxplot(
    data=df,
    x="channel group",
    y="Engagement rate",

    palette="coolwarm"
)
plt.title("Average Engagement Time by Channel")
plt.xticks(rotation=45)
plt.show()
```

# Engaged vs Non Engaged 

```python
session_df = df.groupby("channel group")[["Sessions", "Engaged sessions"]].sum().reset_index()

session_df["Non-Engaged"] = session_df["Sessions"] - session_df["Engaged sessions"]

session_df_melted = session_df.melt(
    id_vars="channel group",
    value_vars=["Engaged sessions", "Non-Engaged"],
    var_name="Session Type",
    value_name="Count"
)

plt.figure(figsize=(9, 7))
sns.barplot(
    data=session_df_melted,
    x="channel group",
    y="Count",
    hue="Session Type"
)
plt.title("Engaged vs Non-Engaged Sessions")
plt.xticks(rotation=45)
plt.show()
```

# Traffic by hour and channel

```python
heatmap_data = df.groupby(["Hour", "channel group"])["Sessions"].sum().unstack().fillna(0)

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap ="YlGnBu",linewidths=.5, annot=True, fmt='0f')
   
plt.title("⏰ Traffic by Hour and Channel")
plt.xlabel("Channel Group")
plt.ylabel("Hour of Day")
plt.show()

```

# engagement rate vs session over time

```python
df_plot = df.groupby("DateHour")[["Engagement rate", "Sessions"]].mean().reset_index()

plt.figure(figsize=(10, 5))

plt.plot(df_plot["DateHour"], df_plot["Engagement rate"], label="Engagement rate", color="green", marker='o')
plt.plot(df_plot["DateHour"], df_plot["Sessions"], label="Sessions", color="blue", marker='s')

plt.title("📊 Engagement Rate and Sessions Over Time")
plt.xlabel("DateHour")
plt.ylabel("Average Value")
plt.legend()
plt.grid(True)
plt.show()
```
